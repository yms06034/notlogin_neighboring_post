```python
import multiprocessing
import time

def your_function():
    while True:
        # 여기에 함수의 내용을 작성합니다.
        print("Function is running...")
        time.sleep(1)

def check_condition(queue):
    print("Checking condition...")
    queue.put("terminate")  # 종료 신호를 Queue에 넣습니다.
```

```python
if __name__ == "__main__":
    # Queue를 생성합니다.
    queue = multiprocessing.Queue()
    
    # 함수를 병렬로 실행합니다.
    process_list = []
    process1 = multiprocessing.Process(target=your_function)
    process_list.append(process1)
    process2 = multiprocessing.Process(target=check_condition, args=(queue,))
    process_list.append(process2)
    
    # 각 프로세스를 시작합니다.
    for process in process_list:
        process.start()
    
    # Queue에서 종료 신호를 확인하고, 종료 조건이 만족되면 모든 프로세스를 종료합니다.
    while True:
        if not queue.empty():
            message = queue.get()
            if message == "terminate":
                print("Terminating processes...")
                for process in process_list:
                    process.terminate()
                break
        time.sleep(1)
    
    # 함수가 종료되었음을 알립니다.
    print("Function has finished.")
```
